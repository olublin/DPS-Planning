```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
#read all of the region files and capacity data

HS_regions = gpd.read_file(r'C:\Users\kevan\OneDrive\Desktop\Data+\HS_regions_2025_26\HS_regions_2025_26.shp')
HS_regions = HS_regions.to_crs('EPSG:4326')

MS_regions = gpd.read_file(r'C:\Users\kevan\OneDrive\Desktop\Data+\MS_regions_2025_26\MS_regions_2025_26.shp')
MS_regions = MS_regions.to_crs('EPSG:4326')

ES_regions = gpd.read_file(r'C:\Users\kevan\OneDrive\Desktop\Data+\ES_regions_2024_25\ES_regions_2024_25.shp')
ES_regions = ES_regions.to_crs('EPSG:4326')

capacity_data = pd.read_csv(r'C:\Users\kevan\OneDrive\Desktop\Data+\Parcels_.csv')
```

```python
#add school name column based on its ID
HS_regions = HS_regions.rename(columns={'sch_id_gt_': 'ID'})
MS_regions = MS_regions.rename(columns={'sch_id_gt_': 'ID'})
ES_regions = ES_regions.rename(columns={'sch_id_gt_': 'ID'})

HS_regions['ID'] = HS_regions['ID'].astype(str)
MS_regions['ID'] = MS_regions['ID'].astype(str)
ES_regions['ID'] = ES_regions['ID'].astype(str)
capacity_data['ID'] = capacity_data['ID'].astype(str)

HS_regions = HS_regions.merge(
    capacity_data[['ID', 'School', 'curr_crowd_assume_100']],
    on='ID',
    how='left'  # Use 'left' to keep all rows from HS_regions
)
MS_regions = MS_regions.merge(
    capacity_data[['ID', 'School', 'curr_crowd_assume_100']],
    on='ID',
    how='left'
)
ES_regions = ES_regions.merge(
    capacity_data[['ID', 'School', 'curr_crowd_assume_100']],
    on='ID',
    how='left'
)
```
